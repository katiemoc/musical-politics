## President Dataset

Now, we will read in our President data.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 

president = pd.read_csv("1976-2020-president.csv")
president
```

Notice that the "notes" column is NaN in every entry. We need to drop this column. 
After, we can drop rows with at least one NaN value because it is either a candidate that was written in or candidate with no party-- both of which we do not need for our project.

```python
# Make a separate DataFrame for tidying
tidy_president = president

# Drop 'notes' column and NaN rows
tidy_president.drop('notes', axis=1, inplace=True)
tidy_president.dropna(inplace=True)
tidy_president
```

```python
# Check for missing data
tidy_president.isna().any(axis=None) 
```

Since the code returns False, we know we have no other missing data in our dataset!

Now we can check the types of all of our columns.

```python
tidy_president.dtypes
```

All of the numeric columns are already ints or floats, and we do not need to change the type of any other column.

We can now start filtering out data that are not relevant to our project-- since we are using Spotify data from 2011 to 2019, we will only use election data from these years as well. Note that this means we will have data strictly from the 2012 and 2016 elections.

```python
# Filter by year
tidy_president = president[(president['year'] >= 2011) & (president['year'] <= 2019)]
tidy_president.head()
```

We also want to compare only the democrat and republican candidates, as they receive the majority of votes during each election.
To be safe, we will strip and make every value uppercase in the "party_simplified" column.

```python
# Filter by party
tidy_president['party_simplified'] = tidy_president['party_simplified'].str.strip().str.upper()
tidy_president = tidy_president[tidy_president['party_simplified'].isin(['DEMOCRAT','REPUBLICAN'])]
tidy_president.head()
```

We are only focusing on swing states' data, so we will clean our data to only include those.
From our description above, the 13 most widely known swing states are Wisconsin, Pennsylvania, New Hampshire, 
Minnesota, Arizona, Georgia, Virginia, Florida, Michigan, Nevada, Colorado, North Carolina, and Maine.
To be safe, we will strip and make every value uppercase in the "state" column.

```python
# Filter by state
tidy_president['state'] = tidy_president['state'].str.strip().str.upper()
tidy_president = tidy_president[tidy_president['state'].isin(['WISCONSIN','PENNSYLVANIA','NEW HAMPSHIRE','MINNESOTA',\
                                                              'ARIZONA','GEORGIA','VIRGINA', 'FLORIDA','MICHIGAN','NEVADA',\
                                                              'COLORADO','NORTH CAROLINA', 'MAINE'])]
tidy_president.head()
```

Lastly, we are going to focus on the "year", "state", "party_simplified", "totalvotes", and "candidatevotes" columns.


```python
# Reset the index
tidy_president = tidy_president.reset_index()

# Filter by specified column name
tidy_president = tidy_president[["year", "state", "party_simplified", "totalvotes", "candidatevotes"]]
tidy_president.head(10)
```

Notice that there are a lot of duplicates in the `state`, `party_simplified`, and `totalvotes` columns. We can simplify this by finding the winning party in each state, using the maximum values in the `candidatevotes` column per year and state.

```python
# Get the winning candidates vote count
winners = tidy_president.groupby(['year','state']).max()['candidatevotes']
winners.head()
```

We can now use the `winners` data to filter our dataset.

```python
# Filter winning candidate rows only
tidy_president = tidy_president[tidy_president['candidatevotes'].isin(winners)]
tidy_president.head()
```

Now the `party_simplified` column tells us the winning party for each year and state! Let's rename this column. Furthermore, we can remove the `totalvotes` and `candidatevotes` columns now that we have this information.

```python
# Change 'party_simplified' column name to 'winning_party'
tidy_president.rename(columns={'party_simplified': 'winning_party'}, inplace=True)

# Drop 'totalvotes' and 'candidatevotes' columns
tidy_president.drop(columns=['totalvotes','candidatevotes'], inplace=True)
tidy_president.head()
```
