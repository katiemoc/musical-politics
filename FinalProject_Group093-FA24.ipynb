Let's see how many different genres are in the data.

```python
genres = tidy_spotify['top genre'].unique()
print(genres)
print(len(genres))
```

The large variety of genres is not ideal for data visualization. We can organize these genres into 10 general categories: `hiphop`,`soul`,`rock`,`pop`,`rap`,`rnb`,`metal`,`dance`,`country`, and `other`. This will help us later on when making graphs and deciphering correlations!

```python
# Function to organize genres into 10 general categories
def standardize_genre(string):
    string = string.lower()
    string = string.strip()
    string = string.replace(" ", "")
    
    if 'hiphop' in string:
        output = 'hiphop'
    elif 'soul' in string:
        output = 'soul'
    elif 'rock' in string:
        output = 'rock'
    elif 'pop' in string:
        output = 'pop'
    elif 'rap' in string:
        output = 'rap'
    elif 'r&b'in string:
        output = 'rnb'
    elif 'metal' in string:
        output = 'metal'
    elif 'dance'  in string:
        output = 'dance'
    elif 'country' in string:
        output = 'country'
    else:
        output = 'other'
    return output

# Apply function
tidy_spotify['top genre'] = tidy_spotify['top genre'].apply(standardize_genre)
tidy_spotify.head()
```

### Hypothesis Test

We would like to know if there is a significant correlation between `genre` and `winning_party` thus let us perform a hypothesis test to prove this, with a significance level of 0.05 and using the Chi-Squared Test to see if the distribution of genres differ between parties.



> Null Hypothesis: `genre` and `winning_party` have NO SIGNIFICANT correlation with one another.

> Alternate Hypothesis: `genre` and `winning_party` have SIGNIFICANT correlation with one another

```python

# Create a contingency table for Chi-Square Test
contingency_table = pd.crosstab(tidy["top genre"], tidy["winning_party"])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output the results
print("Chi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p}")

```

Given our significance level of 0.05 we can see that the resulting P-value is in fact greater than 0.05 and thus we fail to reject the null hypothesis, leading us to the conclusion that `genre` and `winning_party` have no significant correlation between the two. Although this is quite a disappointing result we can try to see if the next column `spch`, which correlates with lyrical density, has any relations with `winning_party`.

### Hypothesis Test

However in statistics an eyeball test is rarely, if ever, accepted as sufficient evidence and thus we will perform a hypothesis test with a significance value of 0.05 to see if there is a correlation between lyrical density and the winning party.

> Null Hypothesis: Lyrical density (`spch`) has NO significant correlation with the winning party (`winning_party`)

> Alternate Hypothesis: Lyrical density (`spch`) DOES significantly correlate with the winning party (`winning_party`)

In this hypothesis test, due to our variables being categorical `winning_party` and numerical `spch` we will conduct this test using the ANOVA test which will test if the mean lyrical density differs significantly across parties. Although the ANOVA test excels at finding significance if the relationship between `spch` and `winning_party` is linear we will also use the Kruskal-Wallis test in case the relationship is non-linear.

```python

# We group the data by winning_party and spch to set up for a hypothesis test
grouped_data = [group["spch"] for _, group in tidy.groupby("winning_party")]

# Checking assumptions using ANOVA
anova_result = f_oneway(*grouped_data)

# Checking asssumptions using Kruskal-Wallis
kruskal_result = kruskal(*grouped_data)

# Output results
print("ANOVA Test Results:")
print(f"F-statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")

print("\nKruskal-Wallis Test Results:")
print(f"Statistic: {kruskal_result.statistic}, p-value: {kruskal_result.pvalue}")
```

Given our results here we can see that the p-value for both the ANOVA test and the Kruskal-Wallis test are greater than 0.05 and thus we can say that we failed to reject the null hypothesis and that there lyrical density does NOT significantly correlate with the winning party.

The Kaggle dataset we are using analyzes US election results from 1976-2020, based on U.S. states, districts, and on the national level. We assume that since these election results are public nationwide that there wouldn’t be problems with the privacy/terms of use of this data. However, the data itself can have bias since voter turnout has varied from year-to-year, which doesn’t account for the public opinion of everyone in a certain state. Generally, more voter turnout has resulted in elections that were more high-stakes compared to other elections. As well, there are laws that prevent certain demographics from voting: for instance, non-citizens, including legal residents, convicted felons, and some with a mental disability cannot vote based on different states. 

The Spotify dataset was made using Spotify’s web API, and does not rely on the data of their individual users’ listening activity in a way it breaches their personal privacy. However, Spotify wasn’t launched in the U.S. until 2011, and after the trial period ended in January 2012 users were limited to ten hours of streaming per month. We cannot analyze the relationship of music preference and U.S. political preference based on Spotify alone for years before 2011 because of this reason. To answer our research question with best accuracy, we must analyze Spotify listening trends from 2011 and beyond in relation to U.S. election results to show a more direct correlation.

We notice that our dataset of Spotify’s top hits seems to contain a large amount of songs in the “pop” genre, along with hip-hop second (21% and 14% respectively). This could be related to the culture surrounding the time, including the values, conflicts, and social/political climate. Generally, pop music is known for its easy-to-remember lyrics, catchy tune, and feel-good factor that makes people feel happier. Radio stations, advertisements, and the public sphere tend to play pop music for the reason of it being uplifting and easily memorable compared to other genres of music. Therefore, the emphasis on pop music in society can overshadow other music genres that other communities resonate with more for its lyrics and values. Since we are excluding Spotify listening trends prior to 2011, our data will primarily pertain to the U.S. population instead of other countries where Spotify had already been released.

Our background research led us to believe that since Democrats typically preferred rap and R&B genres to express songs about their struggles that this would have a visible correlation to states that ended up voting Democratic. As well, research indicated that Republicans did typically prefer country music and Democrats preferred pop, alternative, rap, hip-hop and rock music. Historically, the preference of music taste has indicated a factor of someone’s political affiliation. 

Contrary to our hypothesis, the results of our data analysis demonstrated that even with an increase in the pop, R&B, soul, dance, and uncategorized genres between 2012 and 2016, the swing states that were Democratic in 2012 switched to Republican. The results lead us to remember that correlation does not mean causation, and the reason for swing states switching to Republican may be due to other factors such as campaign success, voter turnout, and other aspects of societal culture. 

In relation to our Background section, where prior research has seen correlations between music genre and political affiliation, our research furthers this by focusing on US swing state data, as these are the states that social and political events can influence. We also explore the relationship between lyrical density and party– despite not finding a statistically significant result, there could be potential if we had more specific data.

As for the available datasets, we were limited to only a few that were relevant to our project. In the ideal case we would have been able to perform our analysis on a Spotify dataset that included the top genres/songs in each state throughout the years to make a clearer distinction between the impact of top genres in each state based on lyrical density and political affiliations. Had we been able to find a dataset that aligned with this limitation we could have furthered our scope towards analyzing individual swing states, most notably those that have been recognized to be toss-ups in recent elections. As we were limited to Spotify data, we would have liked to analyze Apple Music and SoundCloud data as well to have a better understanding of the relationships studied in our project and their overall influence as predictors for swing state outcomes.
